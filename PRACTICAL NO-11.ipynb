```python
import numpy as np
from matplotlib  import pyplot as plt 
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=x*2
```

```python
y
```

```python
plt.plot(x,y)
plt.title("Line Chat")
plt.xlabel("x,axis")
plt.ylabel("y,axis")
plt.show()
```

```python
plt.bar(x,y)
plt.title("Bar Chat")
plt.xlabel("x,axis")
plt.ylabel("y,axis")
plt.show()
```

```python

a = (1,4,7,2,8,3,6)
b = (32,56,78,23,11,22,34)

plt.scatter(a,b)
plt.title("Scatter Chart")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```
